```python
import nibabel.freesurfer.io as fsio
import numpy as np
import os
import tkinter as tk
from tkinter import filedialog
import scipy.io as scio 
import glob
```

```python
root = tk.Tk()
root.lift()
root.withdraw()
file_dir = filedialog.askdirectory(title='FS Subject Directory')
files = glob.glob(file_dir+r'/label/*.HCPMMP1.*')
print(files)
```

```python
for file in files:
    filename = file.split('\\')[-1] + '.mat'
    labels, ctab, names = fsio.read_annot(file)
    label_data = ctab[labels, 4]
    scio.savemat(file_dir + '/' + filename, {"labels_tab" : label_data, "labels": labels, "names" : names })
```
