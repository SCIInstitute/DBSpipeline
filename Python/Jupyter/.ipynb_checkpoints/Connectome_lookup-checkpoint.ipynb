```python
import numpy as np
import pandas as pd
import glob
```

```python
#HCP
with open('Z:\Dropbox (UFL)\DataProcessing\Pipeline Code\Bash\Freesurfer\hcpmmp1_subcortex.txt','r') as f:
    lookup = {}
    labels = []
    index = []
    filename = []
    filepath = []
    for line in f:
        if line.startswith('#') or len(line) < 10:
            continue
        else:
            label_data = line.split()
            labels.append(label_data[1])
            index.append(int(label_data[0]))
            filename.append('HCP_FS.nii.gz')
            filepath.append('HCP')

lookup['Index'] = index
lookup['Labels'] = labels
lookup['File Index'] = index.copy()
lookup['Filename'] = filename
lookup['Path'] = filepath

df_lookup = pd.DataFrame(data=lookup)
#df_lookup.set_index('Index',inplace=True)
df_lookup
```

```python
#Thalamus
thal_names = ['LEFT_AV_2', 
              'LEFT_CL_17', 
              'LEFT_CM_11', 
              'LEFT_HB_13', 
              'LEFT_LGN_9', 
              'LEFT_MDPF_12', 
              'LEFT_MGN_10', 
              'LEFT_MTT_14', 
              'LEFT_PUL_8', 
              'LEFT_THALAMUS_1', 
              'LEFT_VA_4', 
              'LEFT_VLA_5', 
              'LEFT_VLP_6', 
              'LEFT_VPL_7', 
              'LEFT_VPM_18', 
              'RIGHT_AV_2', 
              'RIGHT_CL_17', 
              'RIGHT_CM_11', 
              'RIGHT_HB_13', 
              'RIGHT_LGN_9', 
              'RIGHT_MDPF_12', 
              'RIGHT_MGN_10', 
              'RIGHT_MTT_14', 
              'RIGHT_PUL_8', 
              'RIGHT_THALAMUS_1', 
              'RIGHT_VA_4', 
              'RIGHT_VLA_5', 
              'RIGHT_VLP_6', 
              'RIGHT_VPL_7', 
              'RIGHT_VPM_18']

index = list(range(1000,len(thal_names)+1000))
filename = []
filepath = []
file_ind = []
structure = []
for name in thal_names:
    if 'LEFT' in name:
        filename.append('atlas_CL_left.nii.gz')
    if 'RIGHT' in name:
        filename.append('atlas_CL_right.nii.gz')
    filepath.append('THOMAS')
    file_ind.append(int(name.split('_')[-1]))
    structure.append('Thalamus')

lookup["Index"].extend(index)
lookup["Labels"].extend(thal_names)
lookup['File Index'].extend(file_ind)
lookup['Filename'].extend(filename)
lookup['Path'].extend(filepath)

df_lookup = pd.DataFrame(data=lookup)
#df_lookup.set_index('Index',inplace=True)
```
